$y_{ij} = g_{ij} + j b_{ij}$

$Y_{ij} = G_{ij} + jB_{ij}$

```python
G=zeros([14,14])
B=zeros([14,14])

for x in range(14):
    for y in range(14):
        if x!=y:
            G[x,y]=-3
            B[x,y]=-8
```
